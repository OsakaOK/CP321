## Question 1

Compare the mean Regular Unleaded Gasoline price among different cities in 2022. Plot the corresponding bar chart.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read data
df = pd.read_csv('data/fueltypesall1991-2023.csv')

# Filter the data to only include 2022 and Regular Unleaded Gasoline
gas_type = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') ]
date_gas = gas_type[(df['Date'] >= '2022-01-01') & (df['Date'] < '2023-01-01')]
# Filter the data to only include cities
city = date_gas.loc[:, "Ottawa": "Parry Sound"]
# Calculate the mean retail price for Regular Unleaded Gasoline in each city
all_mean = city.mean()

# Plot the data 
all_mean.plot(kind='bar',color = 'green')

# Add a title and axis labels
plt.title('Mean Retail Price of Regular Unleaded Gasoline in Different Cities in 2022')
plt.xlabel('Cities')
plt.ylabel('Price')

# Show the plot
plt.show()
```

## Question 2

Plot the histogram of the Regular Unleaded Gasoline price in Ottawa for 2022. Choose the appropriate parameters.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read data
df = pd.read_csv('data/fueltypesall1991-2023.csv')

# Filter for only Regular Unleaded Gasoline prices in Ottawa for 2022
gas_type = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') ]
date_gas = gas_type[(df['Date'] >= '2022-01-01') & (df['Date'] < '2023-01-01')]
df_ottawa = date_gas['Ottawa']
# Plot the histogram
plt.hist(df_ottawa, edgecolor = 'green' )
# Add a title and axis labels
plt.title('Histogram of Regular Unleaded Gasoline Prices in Ottawa in 2022')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
